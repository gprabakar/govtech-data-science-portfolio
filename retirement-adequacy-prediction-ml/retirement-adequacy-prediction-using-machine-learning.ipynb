```python
import numpy as np
import pandas as pd

np.random.seed(42)

n = 3000

data = pd.DataFrame({
    "age": np.random.randint(25,65,n),
    "salary": np.random.randint(3000,12000,n),
    "years_worked": np.random.randint(1,40,n),
    "housing_withdrawal": np.random.randint(0,150000,n),
    "voluntary_contribution": np.random.randint(0,20000,n),
    "investment_return_rate": np.random.uniform(0.02,0.09,n)
})

# realistic retirement formula
data["retirement_score"] = (
    data["salary"]*data["years_worked"]*0.15
    + data["voluntary_contribution"]
    - data["housing_withdrawal"]*0.25
    + data["investment_return_rate"]*50000
)

# balanced classification
threshold = data["retirement_score"].median()

data["retirement_adequate"] = (data["retirement_score"] > threshold).astype(int)

data.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

X = data.drop(["retirement_adequate","retirement_score"],axis=1)

y = data["retirement_adequate"]

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)

model = RandomForestClassifier(n_estimators=200)

model.fit(X_train,y_train)

pred = model.predict(X_test)

print(classification_report(y_test,pred))
```

```python
import matplotlib.pyplot as plt

importance = model.feature_importances_

importance_df = pd.DataFrame({
    "feature": X.columns,
    "importance": importance
}).sort_values("importance")

plt.figure(figsize=(8,5))

plt.barh(importance_df["feature"], importance_df["importance"])

plt.title("Retirement Adequacy Drivers")

plt.xlabel("Importance Score")

plt.show()

print(importance_df)
```

```python
import numpy as np
import pandas as pd

np.random.seed(42)

n = 1000

data = pd.DataFrame({
    "age": np.random.randint(18, 80, n),
    "income": np.random.randint(800, 8000, n),
    "employment_status": np.random.choice([0,1], n),  # 0 unemployed, 1 employed
    "household_size": np.random.randint(1, 6, n),
    "education_level": np.random.randint(1, 5, n),
    "medical_condition": np.random.choice([0,1], n),
    "previous_assistance": np.random.choice([0,1], n)
})

# risk score logic
data["risk"] = (
    (data["income"] < 2000).astype(int) +
    (data["employment_status"] == 0).astype(int) +
    data["medical_condition"] +
    data["previous_assistance"]
)

data["high_risk"] = (data["risk"] >= 2).astype(int)

data.to_csv("social_risk_dataset.csv", index=False)

data.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

X = data.drop(["risk","high_risk"], axis=1)
y = data["high_risk"]

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)

model = RandomForestClassifier()
model.fit(X_train, y_train)

pred = model.predict(X_test)

print(classification_report(y_test, pred))
```

```python
import matplotlib.pyplot as plt

importance = model.feature_importances_

plt.barh(X.columns, importance)
plt.title("Risk Factors Importance")
plt.show()
```

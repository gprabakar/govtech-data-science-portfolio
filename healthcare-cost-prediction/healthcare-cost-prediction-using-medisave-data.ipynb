```python
# Step 1 — Dataset
import numpy as np
import pandas as pd
n = 2000

health = pd.DataFrame({

"age": np.random.randint(25,85,n),

"chronic_disease": np.random.choice([0,1],n),

"hospital_visits": np.random.randint(0,12,n),

"insurance_coverage": np.random.randint(0,20000,n),

"health_risk_score": np.random.uniform(0,1,n)

})

health["annual_cost"] = (
health["age"]*20 +
health["chronic_disease"]*4000 +
health["hospital_visits"]*800 +
health["health_risk_score"]*3000 -
health["insurance_coverage"]
)

health.head()
```

```python
# Step 2 — Regression Model

from sklearn.linear_model import LinearRegression

X = health.drop("annual_cost",axis=1)

y = health["annual_cost"]

model = LinearRegression()

model.fit(X,y)

pred = model.predict(X)

print("Model trained successfully")
```

```python
# Step 3 — Plot

import matplotlib.pyplot as plt
plt.scatter(y,pred)

plt.title("Healthcare Cost Prediction")

plt.show()
```

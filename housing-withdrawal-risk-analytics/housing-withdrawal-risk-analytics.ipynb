```python
# Step 1 — Dataset
import numpy as np
import pandas as pd
from sklearn.metrics import classification_report
n = 1800

housing = pd.DataFrame({

"age": np.random.randint(25,60,n),

"salary": np.random.randint(3000,12000,n),

"loan_amount": np.random.randint(50000,600000,n),

"interest_rate": np.random.uniform(0.02,0.05,n),

"years_to_pay": np.random.randint(10,30,n),

"savings_balance": np.random.randint(20000,250000,n)

})

housing["withdrawal_risk"] = (
housing["loan_amount"]/housing["salary"] > 60
).astype(int)

housing.head()
```

```python
# Step 2 — Classification Model

from sklearn.ensemble import GradientBoostingClassifier

X = housing.drop("withdrawal_risk",axis=1)

y = housing["withdrawal_risk"]

model = GradientBoostingClassifier()

model.fit(X,y)

pred = model.predict(X)

print(classification_report(y,pred))
```

```python
# Step 3 — Visualization
import matplotlib.pyplot as plt
housing["withdrawal_risk"].value_counts().plot(kind="bar")

plt.title("Housing Withdrawal Risk Distribution")

plt.show()
```

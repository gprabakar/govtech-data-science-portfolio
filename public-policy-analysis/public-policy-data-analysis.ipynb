```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)

years = np.arange(2000,2025)

policy_data = pd.DataFrame({
    "year": years,
    "education_spending": np.random.randint(5000,15000,len(years)),
    "health_spending": np.random.randint(7000,20000,len(years)),
    "employment_rate": np.random.uniform(85,97,len(years)),
    "life_expectancy": np.random.uniform(72,85,len(years))
})

policy_data.to_csv("policy_dataset.csv", index=False)

policy_data.head()
```

```python
plt.plot(policy_data["year"], policy_data["education_spending"])

plt.title("Education Spending Trend")
plt.xlabel("Year")
plt.ylabel("Spending")

plt.show()
```

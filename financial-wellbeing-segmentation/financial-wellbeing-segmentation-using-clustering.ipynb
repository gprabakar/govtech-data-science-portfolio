```python
# Step 1 — Dataset

import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler

n = 1500

finance = pd.DataFrame({

"savings_balance": np.random.randint(5000,200000,n),

"salary": np.random.randint(2000,12000,n),

"debt": np.random.randint(0,150000,n),

"monthly_expense": np.random.randint(1000,7000,n)

})

finance.head()
```

```python
# Step 2 — Scaling

scaler = StandardScaler()

scaled_data = scaler.fit_transform(finance)
```

```python
# Step 3 — Clustering

from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3)

finance["segment"] = kmeans.fit_predict(scaled_data)

finance.head()
```

```python
# Step 4 — Visualization

import matplotlib.pyplot as plt
plt.scatter(finance["salary"],finance["savings_balance"],c=finance["segment"])

plt.xlabel("Salary")

plt.ylabel("Savings")

plt.title("Financial Wellbeing Segments")

plt.show()
```

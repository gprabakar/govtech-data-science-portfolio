```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report

np.random.seed(42)

n = 1500

case_data = pd.DataFrame({

    "severity_score": np.random.randint(1,10,n),
    "income_level": np.random.randint(800,6000,n),
    "health_flag": np.random.choice([0,1],n),
    "dependents": np.random.randint(0,5,n),
    "previous_cases": np.random.randint(0,4,n)

})

case_data["priority_score"] = (
    case_data["severity_score"] +
    case_data["health_flag"]*2 +
    case_data["dependents"]
)

case_data["priority_level"] = pd.cut(
    case_data["priority_score"],
    bins=[0,5,8,20],
    labels=["Low","Medium","High"]
)

case_data.to_csv("case_priority_dataset.csv", index=False)

case_data.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

case_data["priority_encoded"] = le.fit_transform(case_data["priority_level"])

X = case_data.drop(["priority_score","priority_level","priority_encoded"], axis=1)
y = case_data["priority_encoded"]

from sklearn.ensemble import GradientBoostingClassifier

model = GradientBoostingClassifier()

model.fit(X,y)

pred = model.predict(X)

print(classification_report(y,pred))
```

```python
case_data["priority_level"].value_counts().plot(kind="bar")

plt.title("Case Priority Distribution")

plt.show()
```

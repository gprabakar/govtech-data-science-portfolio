```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)

n = 1200

service = pd.DataFrame({

    "queue_length": np.random.randint(1,60,n),
    "staff_available": np.random.randint(1,10,n),
    "service_type": np.random.randint(1,5,n),
    "hour_of_day": np.random.randint(8,18,n),
    "day_of_week": np.random.randint(1,7,n)

})

service["waiting_time"] = (
    service["queue_length"]*2 -
    service["staff_available"]*3 +
    np.random.normal(0,5,n)
)

service.to_csv("service_optimization_dataset.csv", index=False)

service.head()
```

```python
from sklearn.linear_model import LinearRegression

X = service.drop("waiting_time", axis=1)
y = service["waiting_time"]

model = LinearRegression()

model.fit(X,y)

pred = model.predict(X)

print("Model trained successfully")
```

```python
plt.scatter(y, pred)

plt.xlabel("Actual Waiting Time")
plt.ylabel("Predicted Waiting Time")

plt.title("Prediction Accuracy")

plt.show()
```
